```python
df1_zomato.rename(columns={'approx_cost(for two people)': 'approx_cost_for_two_people'}, inplace=True)
```

```python
df1_zomato['rate'] = df1_zomato['rate'].replace(['-', 'NEW'], pd.NA)
df1_zomato['rate'] = df1_zomato['rate'].str.replace('/5', '', regex=False)
df1_zomato['rate'] = pd.to_numeric(df1_zomato['rate'], errors='coerce')
median_rating = df1_zomato['rate'].median()
df1_zomato['rate'].fillna(median_rating, inplace=True)
```

```python
df1_zomato.rename(columns={
    'approx_cost(for two people)': 'approx_costfor_two_people',
    'listed_in(type)': 'listed_intype',
    'listed_in(city)': 'listed_incity'
}, inplace=True)
```

```python
df1_zomato['approx_costfor_two_people'] = df1_zomato['approx_costfor_two_people'].astype(str).str.replace(',', '')
df1_zomato['approx_costfor_two_people'] = pd.to_numeric(df1_zomato['approx_costfor_two_people'], errors='coerce')
df1_zomato['approx_costfor_two_people'].fillna(df1_zomato['approx_costfor_two_people'].median(), inplace=True)
```

```python
df1_zomato['dish_liked'].fillna("Not Available", inplace=True)
df1_zomato['cuisines'].fillna("Other", inplace=True)
df1_zomato['rest_type'].fillna("Unknown", inplace=True)
```

```python
median_votes = df1_zomato['votes'].median()
df1_zomato['votes'].fillna(median_votes, inplace=True)
```

```python
df1_zomato['online_order'] = df1_zomato['online_order'].map({'Yes': 1, 'No': 0})
df1_zomato['book_table'] = df1_zomato['book_table'].map({'Yes': 1, 'No': 0})
```

```python
df1_zomato['rate'] = df1_zomato['rate'].astype(float)
df1_zomato['votes'] = df1_zomato['votes'].astype(int)
df1_zomato['approx_costfor_two_people'] = df1_zomato['approx_costfor_two_people'].astype(int)
```
